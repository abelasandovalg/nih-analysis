### Foreign Keys

```python
%%bigquery

select count(*) as count from nih_staging.arm_groups ag left join nih_staging.clinical_studies_main csm on ag.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.clinical_results cr left join nih_staging.clinical_studies_main csm on cr.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.collaborators col left join nih_staging.clinical_studies_main csm on col.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.contacts con left join nih_staging.clinical_studies_main csm on con.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.interventions inter left join nih_staging.clinical_studies_main csm on inter.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.location_countries lc left join nih_staging.clinical_studies_main csm on lc.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.locations loc left join nih_staging.clinical_studies_main csm on loc.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.other_outcomes oo left join nih_staging.clinical_studies_main csm on oo.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.primary_outcomes po left join nih_staging.clinical_studies_main csm on po.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.references ref left join nih_staging.clinical_studies_main csm on ref.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.removed_trial_location_countries rtlc left join nih_staging.clinical_studies_main csm on rtlc.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.responsible_parties rp left join nih_staging.clinical_studies_main csm on rp.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.results_reference rr left join nih_staging.clinical_studies_main csm on rr.nct_number = csm.nct_number 
where csm.nct_number is null
union all 
select count(*) as count from nih_staging.secondary_outcomes so left join nih_staging.clinical_studies_main csm on so.nct_number = csm.nct_number 
where csm.nct_number is null
```
