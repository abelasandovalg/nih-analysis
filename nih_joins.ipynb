### Inner Joins 

#### Query was ran to find experimental group types from the arm group table and find correlations to the lead sponsor agency and class from the clinical studies main table. 

```python
%%bigquery

select csm.nct_number, arm_group_label, arm_group_type, description, lead_sponsor_agency, lead_sponsor_agency_class, official_full_name, overall_official_affiliation
from nih_modeled.clinical_studies_main csm 
inner join nih_modeled.arm_groups ag on csm.nct_number = ag.nct_number 
where ag.arm_group_type = "Experimental"
order by ag.serialid limit 12
```

#### Query was ran to find drug intervention types in connection with sponsor agency and class from the clinical studies main table. 

```python
%%bigquery

select csm.nct_number, intervention_type, intervention_name, lead_sponsor_agency, lead_sponsor_agency_class, official_full_name, overall_official_affiliation
from nih_modeled.clinical_studies_main csm 
inner join nih_modeled.interventions inter on csm.nct_number = inter.nct_number 
where inter.intervention_type = "Drug"
order by inter.serialid limit 12
```

#### Query was ran to find any correlation between safety issues from primary outcome and sponsor agency and class from the clinical studies main table. 

```python
%%bigquery

select csm.nct_number, measure, safety_issue, lead_sponsor_agency, lead_sponsor_agency_class, official_full_name, overall_official_affiliation
from nih_modeled.clinical_studies_main csm 
inner join nih_modeled.primary_outcomes po on csm.nct_number = po.nct_number 
where po.safety_issue = True
order by po.serialid limit 12
```

#### Query was ran to find correlations between safety issues from the secondary outcomes table and sponsor agency and class from the clinical studies main table. 

```python
%%bigquery

select csm.nct_number, measure, safety_issue, lead_sponsor_agency, lead_sponsor_agency_class, official_full_name, overall_official_affiliation
from nih_modeled.clinical_studies_main csm 
inner join nih_modeled.secondary_outcomes so on csm.nct_number = so.nct_number 
where so.safety_issue = True
order by so.serialid limit 12
```

### Outer Joins 

#### Query was ran to find correlations between drug intervention type from the interventions table to the collaborator agency and class from the collaborators table. 

```python
%%bigquery

select inter.nct_number, intervention_type, intervention_name, collaborator_agency, collaborator_agency_class
from nih_modeled.interventions inter 
left outer join nih_modeled.collaborators col on inter.nct_number = col.nct_number 
where inter.intervention_type = "Drug"
order by inter.serialid limit 12
```

#### Query was ran to find correlations between eligibility requirements in eligibility table and sponsor type responsible parties. 

```python
%%bigquery

select el.nct_number, eligibility_study_pop, eligibility_sampling_method, eligibility_criteria, eligibility_gender, eligibility_minimum_age, eligibility_maximum_age, eligibility_healthy_volunteers, name_title, organization, investigator_affiliation, investigator_full_name
from nih_modeled.eligibility el 
left outer join nih_modeled.responsible_parties rp on el.nct_number = rp.nct_number 
where rp.type = "Sponsor"
order by rp.serialid limit 12
```
