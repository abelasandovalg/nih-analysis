### NIH_INGEST

```python
%%bigquery

# the following query gathers all experimental arm groups

select * from nih_staging.arm_groups ag
where ag.arm_group_type = "Experimental"
order by nct_number limit 10
```

```python
%%bigquery

# the following query gathers type, title, and time_frame data from secondary data that assessed a safety issue

select type, title, time_frame from nih_staging.clinical_results cr
where cr.type = "Secondary" and cr.safety_issue = true
order by nct_number limit 10
```

```python
%%bigquery

# the following query gathers official_title, overall_status, enrollment, enrollment_type, eligibility_gender, eligibility_minimum_age, 
# condition, and phase from terminated trials that were from HIV Infections

select official_title, overall_status, enrollment, enrollment_type, eligibility_gender, eligibility_minimum_age, condition, phase from nih_staging.clinical_studies_main csm
where csm.overall_status = "Terminated" and csm.condition = "HIV Infections"
order by nct_number limit 10
```

```python
%%bigquery 

# the following query gathers all collaborator information from the US Fed

select * from nih_staging.collaborators col
where col.collaborator_agency_class = "U.S. Fed"
order by nct_number limit 10
```

```python
%%bigquery 

# the following query gathers all information from Eli Lilly and Company 

select * from nih_staging.contacts con
where con.last_name = "Eli Lilly and Company"
order by nct_number limit 10
```

```python
%%bigquery 

# the following query gathers all of the interventions that were placebo drugs 

select * from nih_staging.interventions int
where int.intervention_type = "Drug" and int.intervention_name = "Placebo"
order by nct_number limit 10
```

```python
%%bigquery 

# the following query gathers all information from clinical trials in China 

select * from nih_staging.location_countries locc
where locc.country = "China"
order by nct_number limit 10
```

```python
%%bigquery 

# the following query gathers facility name and city from all NIH clinical trial locations in Houston

select facility_name, facility_city from nih_staging.locations loc
where loc.facility_city = "Houston"
order by nct_number limit 10
```

```python
%%bigquery 

# the following query selects all the data from adverse events that assessed a safety issue

select * from nih_staging.other_outcomes oo
where oo.measure = "Adverse events" and oo.safety_issue = True
order by nct_number limit 10
```

```python
%%bigquery 

# the following query gathers all information from primary outcomes that assessed a measure of safety

select * from nih_staging.primary_outcomes po
where po.measure = "Safety" and po.safety_issue = True
order by nct_number limit 10
```

```python
%%bigquery 

# the following query fathers all references that were not published in scientific literature 

select * from nih_staging.references ref
where ref.citation = "This study has not been published in the scientific literature."
order by nct_number limit 10
```

```python
%%bigquery 

# the following query gathers all of the information from the removed trials locations from the UK

select * from nih_staging.removed_trial_location_countries rtlc
where rtlc.country = "United Kingdom"
order by nct_number limit 10
```

```python
%%bigquery 

# the following query gathers all the data from responsible parties that were affiliated with principal investigators at UT Austin
 
select * from nih_staging.responsible_parties rp
where rp.type = "Principal Investigator" and rp.investigator_affiliation = "University of Texas at Austin"
order by nct_number limit 10
```

```python
%%bigquery

# the following query gathers all reference results from the pmid 21518985 which appears to be from Stien et. al

select * from nih_staging.results_reference rr
where rr.pmid = 21518985
order by nct_number limit 10
```

```python
%%bigquery

# the following query gathers all secondary outcomes that measured adverse events that assesed a safety issue 

select * from nih_staging.secondary_outcomes so
where so.measure = "Adverse events" and so.safety_issue = True
order by nct_number limit 10
```
